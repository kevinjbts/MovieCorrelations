```python
# Importing Packages

import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# Import dataset and clean data

movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
for col in movies.columns:
    pct_missing = np.mean(movies[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
movies.dtypes
```

```python
movies['year_correct'] = movies['released'].astype(str).str[:4]
```

```python
movies['released'] = movies['released'].astype('datetime')
```

```python
movies.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
movies
```

```python
#Drop any duplicates

movies['name'].drop_duplicates().sort_values(ascending=False)
```

```python
# Start understanding the correlation of data: 
# Hypothesis is finding highly correlated items with estiamtes on budget

# Budget high correlation
# Company has correlation

plt.scatter(x=movies['budget'], y=movies['gross'], color='blue')
plt.title('Budget v. Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python


sns.regplot(x='budget', y='gross', data=movies)
```

```python
# looking at correlation across data (note, should only focus on numberical columns)

movies.corr(method='pearson')
```

```python
correlation_matrix = movies.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looks at company

movies.head()
```

```python
# convert string values into numeric utilizing cat codes

df_numerized = movies

for col_name in df_numerized.columns: 
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
    df_numerized
```

```python
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr 
```
